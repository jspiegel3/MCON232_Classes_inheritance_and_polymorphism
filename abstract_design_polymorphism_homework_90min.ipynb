# 1. Three Kinds of Methods

## Instance Methods

Instance methods work with ONE specific object.

Python automatically provides:

```python
self
```

Example:

```python
class Student:

    def __init__(self, name):
        self.name = name

    def show_name(self):
        return self.name
```

Why does `show_name()` need `self`?

Because different objects may have different names.

---

# Class Methods

Class methods work with the CLASS itself.

Python automatically provides:

```python
cls
```

Class methods are often used for:
- counters,
- shared configuration,
- factory methods,
- class-level data.

Example:

```python
class Student:

    student_count = 0

    def __init__(self, name):
        self.name = name
        Student.student_count += 1

    @classmethod
    def get_student_count(cls):
        return cls.student_count
```

Usage:

```python
print(Student.get_student_count())
```

---

# Static Methods

Static methods:
- do NOT need `self`
- do NOT need `cls`

They are usually:
- helpers,
- validators,
- formatters,
- utility methods.

Example:

```python
class GuessHelper:

    @staticmethod
    def clean_text(text: str):
        return text.lower().strip()
```

Usage:

```python
GuessHelper.clean_text(" A ")
```

Important:
We do NOT create an object.

---

# 2. Abstract Classes

An abstract class represents:
- a blueprint,
- a contract,
- or a general category.

It is NOT meant to be instantiated directly.

---

# Importing ABC

```python
from abc import ABC, abstractmethod
```

---

# Example

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def speak(self):
        pass
```

This class says:

> “Every child class MUST implement speak().”

---

# Why Abstract Classes Matter

Abstract classes help:
- organize large systems,
- enforce consistency,
- reduce duplication,
- and guarantee important behavior.

---

# Important Rule

This will FAIL:

```python
animal = Animal()
```

because:
- `Animal` is incomplete,
- `speak()` has not been implemented.

---

# 3. Child Classes and Contracts

A child class must implement ALL abstract methods.

Example:

```python
class Dog(Animal):

    def speak(self):
        print("Woof!")
```

Now:

```python
dog = Dog()
```

works correctly.

---

# 4. Inheritance

Inheritance allows a child class to reuse parent behavior.

Example:

```python
class Animal:

    def eat(self):
        print("Eating")


class Dog(Animal):
    pass
```

Now:

```python
d = Dog()
d.eat()
```

works because:
`Dog` inherited `eat()`.

---

# 5. Overriding

Overriding happens when a child creates its OWN version of a parent method.

Example:

```python
class Animal:

    def speak(self):
        print("Some sound")


class Dog(Animal):

    def speak(self):
        print("Woof!")
```

`Dog` overrides the parent implementation.

---

# 6. Polymorphism

Polymorphism means:
different objects can respond to the SAME method name differently.

Example:

```python
class Dog:

    def speak(self):
        print("Woof!")


class Cat:

    def speak(self):
        print("Meow!")


animals = [Dog(), Cat()]

for animal in animals:
    animal.speak()
```

Output:

```python
Woof!
Meow!
```

Same method call:

```python
speak()
```

Different behavior.

# 7. Sets

Sets are very useful for tracking guesses.

Why?
- no duplicates
- fast lookup
- excellent for membership testing

Example:

```python
guesses = set()

guesses.add("a")
guesses.add("b")
```

Check membership:

```python
"a" in guesses
```

# 8. The `all()` Function

`all()` returns `True`
ONLY if every condition is `True`.

Example:

```python
numbers = [True, True, True]

print(all(numbers))
```

Output:

```python
True
```

---

# Example in a Word Game

```python
return all(letter in self.guesses for letter in self.secret_word)
```

Meaning:

> “Has every letter in the secret word been guessed?”

# 11. Design Thinking

A parent class answers:

> “What must all subclasses be able to do?”

Child classes answer:

> “How will I do it?”

That separation is one of the major strengths of object-oriented design.
