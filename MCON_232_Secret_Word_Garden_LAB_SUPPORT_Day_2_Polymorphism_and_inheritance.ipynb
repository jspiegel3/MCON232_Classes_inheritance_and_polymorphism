```python
"""
Insert in to file garden_view.py

VIEW CLASS for The Secret Garden Word Game.

GardenView is responsible for turtle graphics and simple player input.

Important UI note:
    Turtle's screen.textinput() popup appears in the center of the screen.
    Turtle does not give us a simple way to move that popup.

    Because of that, this version moves the game layout slightly downward
    and keeps the main word/message area above the popup as much as possible.

MVC Responsibilities:
    Model:      WordPuzzle stores word logic.
    View:       GardenView draws the screen and asks for guesses.
    Controller: SecretGardenGame coordinates the game.
"""

from turtle import Screen, Turtle


class GardenView:
    """
    GardenView handles all turtle graphics that are not part of a specific Flower.

    Responsibilities:
        1. Create and configure the screen.
        2. Draw the window box.
        3. Draw the six flower slots.
        4. Display the secret word blanks.
        5. Display messages.
        6. Ask the player for a letter guess.
        7. Place flowers in the garden.
    """

    def __init__(self):
        """
        Create the view.

        This prepares:
            - the turtle screen
            - a turtle for drawing the box
            - a turtle for writing text
        """

        self.screen = Screen()
        self.screen.title("The Secret Garden Word Game")

        # Soft garden-sky background.
        self.screen.bgcolor("#48c7d8")
        self.screen.setup(width=900, height=700)

        # tracer(0) turns off automatic animation.
        # We call screen.update() when we want the screen to refresh.
        self.screen.tracer(0)

        # Turtle used for drawing the flower box and slots.
        self.box_turtle = Turtle()
        self.box_turtle.hideturtle()
        self.box_turtle.speed(0)
        self.box_turtle.penup()

        # Turtle used for writing the word, messages, and labels.
        self.text_turtle = Turtle()
        self.text_turtle.hideturtle()
        self.text_turtle.speed(0)
        self.text_turtle.penup()

        # Store slot positions.
        # Each wrong guess will use the next available slot.
        self.slot_positions = []

        self.setup_screen()

    def setup_screen(self):
        """
        Draw the initial screen.

        This method is called once at the beginning.
        """

        self.draw_title()
        self.draw_window_box()
        self.draw_flower_box(max_slots=6)
        self.update_word_display("_ _ _ _ _")
        self.update_wrong_letters_display([])
        self.show_message("Welcome to The Secret Garden Word Game.")
        self.screen.update()

    def draw_title(self):
        """
        Draw the title at the top of the screen.

        The title stays high enough to feel like a header,
        but not so high that it crowds the top of the window.
        """

        self.text_turtle.penup()
        self.text_turtle.goto(0, 305)
        self.text_turtle.color("darkgreen")
        self.text_turtle.write(
            "The Secret Garden Word Game",
            align="center",
            font=("Georgia", 26, "bold")
        )

    def _draw_panel(self, left_x: int, bottom_y: int, width: int, height: int):
        """
        Draw a cream-colored rectangle behind text.

        This helper keeps our text readable against the blue background.

        Parameters:
            left_x (int): left side of the rectangle
            bottom_y (int): bottom of the rectangle
            width (int): rectangle width
            height (int): rectangle height
        """

        self.text_turtle.penup()
        self.text_turtle.goto(left_x, bottom_y)
        self.text_turtle.color("#fffdf5")
        self.text_turtle.begin_fill()
        self.text_turtle.pendown()

        for _ in range(2):
            self.text_turtle.forward(width)
            self.text_turtle.left(90)
            self.text_turtle.forward(height)
            self.text_turtle.left(90)

        self.text_turtle.end_fill()
        self.text_turtle.penup()

    def draw_window_box(self):
        """
        Draw a large window-box planter.

        The box represents the player's six wrong-guess chances.
        Each wrong guess grows one flower in the box.

        This version places the box lower so the turtle text input popup
        has more room in the middle of the screen.
        """

        t = self.box_turtle

        # Main planter outline.
        t.penup()
        t.goto(-360, -300)
        t.color("sienna")
        t.pensize(4)
        t.pendown()

        for _ in range(2):
            t.forward(720)
            t.left(90)
            t.forward(110)
            t.left(90)

        t.penup()

        # Soil fill at the bottom of the planter.
        t.goto(-355, -295)
        t.color("#8B5A2B")
        t.begin_fill()
        t.pendown()

        for _ in range(2):
            t.forward(710)
            t.left(90)
            t.forward(35)
            t.left(90)

        t.end_fill()
        t.penup()

        # Flower-box label.
        self.text_turtle.goto(0, -335)
        self.text_turtle.color("sienna")
        self.text_turtle.write(
            "6 wrong guesses fill the flower box",
            align="center",
            font=("Arial", 14, "bold")
        )

    def draw_flower_box(self, max_slots: int):
        """
        Draw empty flower slots inside the flower box.

        Parameter:
            max_slots (int):
                Number of wrong guesses allowed.

        The slots are low enough that even taller flowers have room to grow
        without covering the word display.
        """

        self.slot_positions.clear()

        start_x = -300
        spacing = 120

        # This y-position is the base of each flower stem.
        # Lower value = flowers start lower.
        y = -265

        for i in range(max_slots):
            x = start_x + i * spacing
            self.slot_positions.append((x, y))

            # Draw an empty slot as a small circle.
            self.box_turtle.penup()
            self.box_turtle.goto(x, y)
            self.box_turtle.color("darkolivegreen")
            self.box_turtle.pensize(2)
            self.box_turtle.pendown()
            self.box_turtle.circle(18)
            self.box_turtle.penup()

            # Slot number.
            self.text_turtle.goto(x, y - 30)
            self.text_turtle.color("darkolivegreen")
            self.text_turtle.write(
                str(i + 1),
                align="center",
                font=("Arial", 10, "normal")
            )

    def update_word_display(self, display_word: str):
        """
        Update the secret word display.

        Parameter:
            display_word (str):
                A string such as "_ E _ R E T".

        This method clears and redraws only the word/message panel area.
        """

        # Draw panel background.
        self._draw_panel(-365, 145, 730, 110)

        # Draw display word.
        self.text_turtle.goto(0, 170)
        self.text_turtle.color("black")
        self.text_turtle.write(
            display_word,
            align="center",
            font=("Courier", 34, "bold")
        )

        self.screen.update()

    def update_wrong_letters_display(self, wrong_letters: list):
        """
        Display wrong letters as a simple text list.

        The wrong letters also appear inside the flowers,
        but this line makes the game easier to read.
        """

        # Draw smaller panel background.
        self._draw_panel(-365, 80, 730, 45)

        if wrong_letters:
            wrong_text = "Wrong guesses: " + ", ".join(wrong_letters)
        else:
            wrong_text = "Wrong guesses:"

        self.text_turtle.goto(0, 94)
        self.text_turtle.color("darkred")
        self.text_turtle.write(
            wrong_text,
            align="center",
            font=("Arial", 14, "bold")
        )

        self.screen.update()

    def show_message(self, message: str):
        """
        Display a message to the player.

        Example messages:
            - "Correct!"
            - "That letter is not in the word."
            - "You win!"

        The message is drawn inside the same upper panel as the word,
        above the blanks.
        """

        # Only erase/redraw the top strip of the word panel.
        self._draw_panel(-365, 215, 730, 40)

        self.text_turtle.goto(0, 228)
        self.text_turtle.color("darkgreen")
        self.text_turtle.write(
            message,
            align="center",
            font=("Arial", 15, "bold")
        )

        self.screen.update()

    def get_letter_guess(self) -> str:
        """
        Ask the player for one letter using a turtle text input box.

        Returns:
            The player's input as a lowercase string.

        If the player cancels, return an empty string.

        Important:
            Turtle places this input box in the center of the screen.
            We cannot move it using turtle commands.
        """

        guess = self.screen.textinput(
            "Make a Guess",
            "Enter one letter:"
        )

        if guess is None:
            return ""

        return guess.lower().strip()

    def get_slot_position(self, index: int):
        """
        Return the x, y position of a flower slot.

        Parameter:
            index (int):
                The wrong guess number, starting at 0.

        Example:
            index 0 returns the first slot.
            index 1 returns the second slot.
        """

        return self.slot_positions[index]

    def place_flower(self, flower):
        """
        Place a flower in the garden.

        Parameter:
            flower:
                Any object that has a draw() method.

        Teaching point:
            GardenView does not need to know whether the flower is a Daisy,
            Tulip, Hydrangea, Bluebell, etc.

            It simply says:
                flower.draw()

            That is polymorphism.
        """

        flower.draw()
        self.screen.update()

    def close_when_clicked(self):
        """
        Keep the turtle window open until the user clicks it.
        """

        self.screen.exitonclick()
```
